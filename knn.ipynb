```python
import numpy as np
import pandas as pd
from collections import Counter

def euclidean_distance(vec1,vec2):
    return np.sqrt(np.sum(np.square(vec1 - vec2)))
    

train_data = {'movie1':[45,2,9,'Love'],
              'movie2':[21,17,5,'Comedy'],
              'movie3':[54,9,11,'Comedy'],
              'movie4':[39,0,31,'Comedy'],
              'movie5':[5,2,57,'Action'],
              'movie6':[3,2,65,'Action'],
              'movie7':[6,4,21,'Action'],
              'movie8':[7,46,4,'Love'],
              'movie9':[9,39,8,'Love'],
              'movie10':[9,38,2,'Love'],
              'movie11':[8,34,17,'Love'],
              'movie12':[2,3,55,'Action']
              }

train_df = pd.DataFrame(train_data).T
train_df.columns = ['Laugh','Hug','Fight','Movie_Type']
test_data = {'movie_test':[23,3,17]} 

K = 4
movie = 'movie_test'

distance_list = []
for train_X in train_df.values[:,:-1]:
    test_X = np.array(test_data[movie])
    distance_list.append(euclidean_distance(train_X,test_X))
    
distance_df = pd.DataFrame({"Distance":distance_list},index=train_df.index)
result = pd.concat([train_df,distance_df],axis=1).sort_values(by="Distance")


d = Counter(result.head(K)['Movie_Type'])
print(d)
print(movie,max(d,key=d.get))
```
